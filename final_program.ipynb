```python
from collections import defaultdict
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import brier_score_loss
from sklearn.metrics import accuracy_score, classification_report

features = [
    'LogRankDiff',      # ranking
    'DiffWR',           # diferencia de winrate históric en herva
    'DiffForm',         # últims 10
    'H2HDiff',          # head-to-head
    'OddsProbDiff',     # diferencia de probabilitat implícita de quotas
    'DiffRest',         # dies de rest
    'EfficiencyDiff',   # winrate / rank → jugador que rendeix per sobre del seu ranking
    'RankDiff'          # diferencia de ranking
]

X_train = train[features].fillna(0)
X_test = first_round[features].fillna(0)
y_train = train['WinnerBinary']
y_test = first_round['WinnerBinary']

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

base_model = XGBClassifier(
    n_estimators=500,                    # Nombre total d'arbres (boosting rounds) que es construiran. 500 és un valor força alt, bo per a un bon rendiment si es controla l'overfitting.
    max_depth=6,                         # Profunditat màxima de cada arbre. 6 és un valor moderat: permet interaccions complexes sense fer els arbres excessivament profunds.
    learning_rate=0.02,                  # Taxa d'aprenentatge (eta). 0.02 és bastant baixa → aprenentatge lent i estable, ideal quan tens molts estimators (500).
    subsample=0.85,                      # Percentatge de mostres (files) que s'utilitzen per entrenar cada arbre. 0.85 = 85% → ajuda a reduir overfitting i afegeix variància (com bagging).
    colsample_bytree=0.85,               # Percentatge de columnes (variables) que s'agafen aleatòriament per cada arbre. 85% → també redueix overfitting i millora la generalització.
    min_child_weight=5,                  # Suma mínima del pes Hessiana en un node fill. Valors >1 fan el model més conservador: evita dividir nodes amb poca informació.
    gamma=0.1,                           # Minimització mínima de la funció de pèrdua necessària per fer una partició addicional (regularització per poda). 0.1 afavoreix arbres més simples.
    reg_alpha=0.1,                       # Terme de regularització L1 sobre els pesos de les fulles. Ajuda a fer sparseness i és útil quan hi ha moltes variables irrellevants.
    reg_lambda=1.0,                      # Terme de regularització L2 (més suau que L1). 1.0 és el valor per defecte i sol funcionar bé en la majoria de casos.
    random_state=42,                     # Llavor per a la reproductibilitat dels resultats (aleatorietat controlada).
    n_jobs=-1,                           # Utilitza tots els nuclis disponibles del processador per entrenar en paral·lel → molt més ràpid.
    eval_metric='logloss'                # Mètrica d'avaluació durant l'entrenament (per classificació binària/multiclasse). 'logloss' = logarithmic loss (cross-entropy).
)

model = CalibratedClassifierCV(base_model, method='isotonic', cv=5)
model.fit(X_train_sc, y_train)

pred = model.predict(X_test_sc)
prob = model.predict_proba(X_test_sc)[:, 1]
acc = accuracy_score(y_test, pred)
brier = brier_score_loss(y_test, prob)

print("\n" + "="*80)
print(f"ACCURACY FINAL TEST CRONOLÓGIC: {acc:.4f}")
print(f"Brier Score: {brier:.4f}")
print(classification_report(y_test, pred, digits=4))
print("="*80)
```

```python
# Time Series Cross-Validation Function
from sklearn.model_selection import TimeSeriesSplit
import warnings
warnings.filterwarnings('ignore')

def time_series_cross_validation(
    train_data,
    features,
    round_name="Round",
    test_data=None,
    n_splits=5,
    model_params=None,
    verbose=True
):
    """
    Perform Time Series Cross-Validation on training data.
    
    Parameters:
    -----------
    train_data : pd.DataFrame
        Training dataframe with 'Date' and 'WinnerBinary' columns
    features : list
        List of feature column names to use
    round_name : str, default="Round"
        Name of the round (for display purposes)
    test_data : pd.DataFrame, optional
        Test data (for display purposes only)
    n_splits : int, default=5
        Number of cross-validation folds
    model_params : dict, optional
        Dictionary of XGBoost parameters. If None, uses default parameters:
        {
            'n_estimators': 500,
            'max_depth': 6,
            'learning_rate': 0.02,
            'subsample': 0.85,
            'colsample_bytree': 0.85,
            'min_child_weight': 5,
            'gamma': 0.1,
            'reg_alpha': 0.1,
            'reg_lambda': 1.0,
            'random_state': 42,
            'n_jobs': -1,
            'eval_metric': 'logloss'
        }
    verbose : bool, default=True
        Whether to print detailed results
    
    Returns:
    --------
    dict : Dictionary containing all CV scores and summary statistics
    """
    # Default model parameters
    default_params = {
        'n_estimators': 500,
        'max_depth': 6,
        'learning_rate': 0.02,
        'subsample': 0.85,
        'colsample_bytree': 0.85,
        'min_child_weight': 5,
        'gamma': 0.1,
        'reg_alpha': 0.1,
        'reg_lambda': 1.0,
        'random_state': 42,
        'n_jobs': -1,
        'eval_metric': 'logloss'
    }
    
    # Merge with user-provided parameters
    if model_params is not None:
        default_params.update(model_params)
    
    # Sort train data by Date to ensure chronological order
    train_sorted = train_data.sort_values('Date').reset_index(drop=True)
    
    # Prepare features
    X_train_full = train_sorted[features].fillna(0)
    y_train_full = train_sorted['WinnerBinary']
    
    # Time Series Cross-Validation with expanding window
    tscv = TimeSeriesSplit(n_splits=n_splits)
    
    cv_scores = {
        'accuracy': [],
        'brier_score': [],
        'precision_0': [],
        'recall_0': [],
        'f1_0': [],
        'precision_1': [],
        'recall_1': [],
        'f1_1': []
    }
    
    if verbose:
        print("="*80)
        print(f"TIME SERIES CROSS-VALIDATION RESULTS - {round_name}")
        print("="*80)
        print(f"\nNumber of folds: {n_splits}")
        print(f"Total training samples: {len(train_sorted)}")
        if test_data is not None:
            print(f"Test set size: {len(test_data)}")
        print("\n" + "-"*80)
    
    for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train_full), 1):
        # Split data
        X_train_fold = X_train_full.iloc[train_idx]
        X_val_fold = X_train_full.iloc[val_idx]
        y_train_fold = y_train_full.iloc[train_idx]
        y_val_fold = y_train_full.iloc[val_idx]
        
        # Scale features
        scaler_fold = StandardScaler()
        X_train_fold_sc = scaler_fold.fit_transform(X_train_fold)
        X_val_fold_sc = scaler_fold.transform(X_val_fold)
        
        # Train model
        base_model_fold = XGBClassifier(**default_params)
        model_fold = CalibratedClassifierCV(base_model_fold, method='isotonic', cv=5)
        model_fold.fit(X_train_fold_sc, y_train_fold)
        
        # Predict
        pred_fold = model_fold.predict(X_val_fold_sc)
        prob_fold = model_fold.predict_proba(X_val_fold_sc)[:, 1]
        
        # Calculate metrics
        acc_fold = accuracy_score(y_val_fold, pred_fold)
        brier_fold = brier_score_loss(y_val_fold, prob_fold)
        report_fold = classification_report(y_val_fold, pred_fold, output_dict=True, zero_division=0)
        
        # Store scores
        cv_scores['accuracy'].append(acc_fold)
        cv_scores['brier_score'].append(brier_fold)
        cv_scores['precision_0'].append(report_fold['0']['precision'])
        cv_scores['recall_0'].append(report_fold['0']['recall'])
        cv_scores['f1_0'].append(report_fold['0']['f1-score'])
        cv_scores['precision_1'].append(report_fold['1']['precision'])
        cv_scores['recall_1'].append(report_fold['1']['recall'])
        cv_scores['f1_1'].append(report_fold['1']['f1-score'])
        
        if verbose:
            # Print fold results
            print(f"\nFold {fold}:")
            print(f"  Train size: {len(train_idx)}, Validation size: {len(val_idx)}")
            print(f"  Accuracy: {acc_fold:.4f}")
            print(f"  Brier Score: {brier_fold:.4f}")
            print(f"  Precision (Class 0): {report_fold['0']['precision']:.4f}")
            print(f"  Recall (Class 0): {report_fold['0']['recall']:.4f}")
            print(f"  F1-Score (Class 0): {report_fold['0']['f1-score']:.4f}")
            print(f"  Precision (Class 1): {report_fold['1']['precision']:.4f}")
            print(f"  Recall (Class 1): {report_fold['1']['recall']:.4f}")
            print(f"  F1-Score (Class 1): {report_fold['1']['f1-score']:.4f}")
    
    # Calculate summary statistics
    summary = {
        'mean_accuracy': np.mean(cv_scores['accuracy']),
        'std_accuracy': np.std(cv_scores['accuracy']),
        'mean_brier': np.mean(cv_scores['brier_score']),
        'std_brier': np.std(cv_scores['brier_score']),
        'mean_precision_0': np.mean(cv_scores['precision_0']),
        'std_precision_0': np.std(cv_scores['precision_0']),
        'mean_recall_0': np.mean(cv_scores['recall_0']),
        'std_recall_0': np.std(cv_scores['recall_0']),
        'mean_f1_0': np.mean(cv_scores['f1_0']),
        'std_f1_0': np.std(cv_scores['f1_0']),
        'mean_precision_1': np.mean(cv_scores['precision_1']),
        'std_precision_1': np.std(cv_scores['precision_1']),
        'mean_recall_1': np.mean(cv_scores['recall_1']),
        'std_recall_1': np.std(cv_scores['recall_1']),
        'mean_f1_1': np.mean(cv_scores['f1_1']),
        'std_f1_1': np.std(cv_scores['f1_1']),
        'all_scores': cv_scores
    }
    
    if verbose:
        # Print summary statistics
        print("\n" + "="*80)
        print("CROSS-VALIDATION SUMMARY")
        print("="*80)
        print(f"\nMean Accuracy: {summary['mean_accuracy']:.4f} (+/- {summary['std_accuracy']:.4f})")
        print(f"Mean Brier Score: {summary['mean_brier']:.4f} (+/- {summary['std_brier']:.4f})")
        print(f"\nClass 0 (Player A wins):")
        print(f"  Mean Precision: {summary['mean_precision_0']:.4f} (+/- {summary['std_precision_0']:.4f})")
        print(f"  Mean Recall: {summary['mean_recall_0']:.4f} (+/- {summary['std_recall_0']:.4f})")
        print(f"  Mean F1-Score: {summary['mean_f1_0']:.4f} (+/- {summary['std_f1_0']:.4f})")
        print(f"\nClass 1 (Player B wins):")
        print(f"  Mean Precision: {summary['mean_precision_1']:.4f} (+/- {summary['std_precision_1']:.4f})")
        print(f"  Mean Recall: {summary['mean_recall_1']:.4f} (+/- {summary['std_recall_1']:.4f})")
        print(f"  Mean F1-Score: {summary['mean_f1_1']:.4f} (+/- {summary['std_f1_1']:.4f})")
        print("="*80)
    
    return summary

```

```python
# Resultats detallats per fila
results_df = first_round[['Date', 'PlayerA', 'PlayerB', 'WinnerBinary']].copy()
results_df['Predicted'] = pred
results_df['Probability'] = prob
results_df['Correct'] = (results_df['WinnerBinary'] == results_df['Predicted'])
results_df['Result'] = results_df['Correct'].map({True: '✓ Correct', False: '✗ Failed'})

# Mostrar resultats
print("RESULTADOS DETALLADOS POR FILA:")
print("="*100)
print(f"\nTotal de predicciones: {len(results_df)}")
print(f"Correctas: {results_df['Correct'].sum()} ({results_df['Correct'].mean()*100:.2f}%)")
print(f"Fallidas: {(~results_df['Correct']).sum()} ({(~results_df['Correct']).mean()*100:.2f}%)")
print("\n" + "="*100 + "\n")

# Mostrar totes les files
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)

display(results_df[['Date', 'PlayerA', 'PlayerB', 'WinnerBinary', 'Predicted', 'Probability', 'Result']])

# Mostrar només les fallides per anàlisis
print("\n" + "="*100)
print("PREDICCIONS FALLIDES:")
print("="*100)
failed = results_df[~results_df['Correct']]
if len(failed) > 0:
    display(failed[['Date', 'PlayerA', 'PlayerB', 'WinnerBinary', 'Predicted', 'Probability']])
else:
    print("Totes les prediccions han sigut correctes!")
```
